```python
# Import core Keras modules from TensorFlow
from keras import Input
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

# For visualization
import matplotlib.pyplot as plt

# For inline graph display inside Jupyter
%matplotlib inline
```

```python
#loading dataset
train_set = "dataset/train"
test_set = "dataset/test"
```

```python
# Initialize CNN
cnn_model = Sequential()

# First Convolution + Pooling layer
cnn_model.add(Input(shape=(64, 64, 3)))
cnn_model.add(Conv2D(32, (3, 3), activation='relu'))
cnn_model.add(MaxPooling2D(pool_size=(2, 2)))

#second layer + pooling layer
cnn_model.add(Conv2D(64, (3, 3), activation='relu'))
cnn_model.add(MaxPooling2D(pool_size=(2, 2)))

# Flattening
cnn_model.add(Flatten())

# Full connection layer
cnn_model.add(Dense(128, activation='relu'))
cnn_model.add(Dropout(0.5))
cnn_model.add(Dense(1, activation='sigmoid'))

#compiling the cnn model
cnn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from keras.callbacks import CSVLogger
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from keras.callbacks import CSVLogger

csv_logger = '../logs/log.csv'

Data_train = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    vertical_flip=True
)

Data_test = ImageDataGenerator(
    rescale=1./255
)

training_set = Data_train.flow_from_directory(
    '../dataset/train',
    target_size=(64,64),
    batch_size=32,
    class_mode='binary'
)

test_set = Data_test.flow_from_directory(
    '../dataset/test',
    target_size=(64,64),
    batch_size=32,
    class_mode='binary'
)

history = cnn_model.fit(
    training_set,
    epochs=2,
    validation_data=test_set,
    callbacks=[
        CSVLogger(csv_logger, append=False, separator=';')
    ]
)

cnn_model.save('../models/cnn_model3.keras')
```

```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt

# Load model
cnn_model = load_model('../models/cnn_model3.keras')

# Load image
img_path = '../dataset/singleImage/new1.jpg'
img = image.load_img(img_path, target_size=(64, 64))

# Show image
plt.imshow(img)
plt.axis('off')
plt.show()

# Preprocess
test_image = image.img_to_array(img)
test_image = np.expand_dims(test_image, axis=0)
test_image = test_image / 255.0

# Predict
result = cnn_model.predict(test_image)

print("Prediction Score:", result[0][0])

if result[0][0] > 0.5:
    prediction = 'Dog'
else:
    prediction = 'Cat'

print("Prediction:", prediction)
```
